## **OpenCV and YOLOv5**

```python
from IPython import get_ipython
from IPython.display import display
# %%
import cv2
from yolo_predictions import YOLO_Pred
from google.colab.patches import cv2_imshow
```

```python
import os

# List all files and directories in the current working directory
files = os.listdir()

# Print the list of files
print(files)
```

```python
#!pip install onnx
import onnx

try:
    model = onnx.load(files[5])  # Load the ONNX model
    onnx.checker.check_model(model)  # Check the model for errors
    print("ONNX model is valid.")
except Exception as e:
    print(f"Error validating ONNX model: {e}")
```

```python
yolo = YOLO_Pred(files[5],'data.yaml')
```

```python
img = cv2.imread(files[1])

cv2_imshow(img) # Use cv2_imshow instead of cv2.imshow
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# predictions
img_pred = yolo.predictions(img)
```

```python

# predictions
img_pred = yolo.predictions(img)

# Use cv2_imshow instead of cv2.imshow to display the image
cv2_imshow(img_pred)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
